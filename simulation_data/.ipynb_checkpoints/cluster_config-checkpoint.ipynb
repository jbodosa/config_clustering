```python
trj_list=[]
for i in np.arange(20,41):
    trj_list.append(f"step5_{i}.dcd")

u = mda.Universe("step3_input.psf",  trj_list)
```

```python
dist_cutoff = 4.3

be = u.select_atoms("resname BE")
glp2_O2L = u.select_atoms("resname GLP2 and type PL")
```

```python
dist_matrix = distance_array(be.positions, glp2_O2L.positions)
```

```python
dist_matrix
```

```python
pairs = np.argwhere(dist_matrix < dist_cutoff)
```

```python

frame_list = []
pair_ids_list = []
labels_list = []

for ts in u.trajectory:
    dist_matrix = distance_array(be.positions, glp2_O2L.positions)
    pairs = np.argwhere(dist_matrix < dist_cutoff)
    midpoints = []
    pair_ids = []
    for i, j in pairs:
        pos_be = be[i].position
        pos_o2l = glp2_O2L[j].position
        midpoint = 0.5 *(pos_be + pos_o2l)
        midpoints.append(midpoint)
        pair_ids.append((i,j))
        
    midpoints = np.array(midpoints)

    if len(midpoints) !=0 :
        db = DBSCAN(eps=5, min_samples=2, metric='euclidean').fit(midpoints)
        labels = db.labels_

        frame_list.append(ts.frame)
        pair_ids_list.append(pair_ids)
        labels_list.append(labels)

frame_dict = pd.DataFrame({"frame": frame_list, "pair_ids": pair_ids_list, "labels": labels_list})
```

```python
def group_pairs_chained_by_position(pairs):
    """
    Group pairs transitively where connections are only formed when
    pair[0] == other_pair[0] OR pair[1] == other_pair[1].
    Returns list of groups; each group is a list of pairs.
    """
    # build maps from first / second element to list of pair indices
    first_map = defaultdict(list)
    second_map = defaultdict(list)
    for idx, (a, b) in enumerate(pairs):
        first_map[a].append(idx)
        second_map[b].append(idx)

    visited = set()
    groups = []

    for i in range(len(pairs)):
        if i in visited:
            continue
        queue = deque([i])
        comp_indices = []
        while queue:
            cur = queue.popleft()
            if cur in visited:
                continue
            visited.add(cur)
            comp_indices.append(cur)
            a, b = pairs[cur]
            # enqueue all pairs sharing same pair[0]
            for neigh in first_map[a]:
                if neigh not in visited:
                    queue.append(neigh)
            # enqueue all pairs sharing same pair[1]
            for neigh in second_map[b]:
                if neigh not in visited:
                    queue.append(neigh)
            # clear lists so we don't re-scan them again
            first_map[a] = []
            second_map[b] = []

        groups.append([pairs[idx] for idx in sorted(comp_indices)])
    return groups

```

```python
# Example usage with your list:
group_frame_dict = {}

for row in range(frame_dict["frame"].shape[0]):
    frame = frame_dict["frame"].iloc[row]
    pairs = frame_dict["pair_ids"].iloc[row]

    groups = group_pairs_chained_by_position(pairs)
    for i, g in enumerate(groups, 1):
        # print(f"Frame: {frame}, Group {i}: {g}")
        g_key = tuple(g)
        if g_key not in group_frame_dict.keys():
            group_frame_dict[g_key] = [frame]
        else:
            # print("Found repeat")
            group_frame_dict[g_key].append(frame)
```

```python
def remove_last_occurrence(text, word):
    parts = text.rsplit(word, 1)  # split from the right just once
    return ''.join(parts)
```

```python

be_1_list = []
be_2_list = []
gnum_dict = {}

for i in range(group_frame_df_cutoff.shape[0]):
    group = group_frame_df_cutoff["group"].iloc[i]
    frame_list = group_frame_df_cutoff["frames"].iloc[i]

    # --- Build selection strings ---
    be_sel = "resname BE and ("
    glp2_sel = "resname GLP2 and ("
    
    for pair in group:
        be_sel   += f" resid {pair[0] + 1} or "
        glp2_sel += f" resid {pair[1] + 1} or "

    be_sel   = remove_last_occurrence(be_sel, "or")   + ")"
    glp2_sel = remove_last_occurrence(glp2_sel, "or") + ")"

    # --- Atom selections ---
    be_atoms = u.select_atoms(be_sel)
    glp2_atoms = u.select_atoms(glp2_sel)
    total_sel = glp2_atoms + be_atoms

    # --- Work on a copy ---
    sel_copy = mda.Merge(total_sel.atoms)

    # --- Reset resids + segid ---
    for seg in sel_copy.segments:
        n_res = len(seg.residues)        
        seg.residues.resids = np.arange(1, n_res + 1)  # reset resids to start at 1
        # seg.segid = seg.residues.resnames[0]           # set segid to residue name

        if seg.residues.resnames[0] == "BE":
            if n_res == 1:
                be_1_list.append(str(i + 1))
            else:
                be_2_list.append(str(i + 1))
        elif seg.residues.resnames[0] == "GLP2":
            gnum_dict[str(i + 1)] = n_res

    # --- Write all frames for this group ---
    with mda.Writer(f"group_{i + 1}.pdb", total_sel.n_atoms) as w:
        for ts in u.trajectory[frame_list]:
            frame_copy = mda.Merge(total_sel.atoms)

            # Reset resids + segid
            for seg in frame_copy.segments:
                n_res = len(seg.residues)
                for idx, res in enumerate(seg.residues, 1):
                    res.resid = idx
                for atom in seg.atoms:
                    atom.segment.segid = seg.residues[0].resname
                    
            center = frame_copy.atoms.center_of_mass()
            frame_copy.atoms.positions -= center  # recenter to (0,0,0)
            w.write(frame_copy)

    # --- Write last frame separately ---
    u.trajectory[frame_list[-1]]
    last_copy = mda.Merge(total_sel.atoms)
    for seg in last_copy.segments:
        for idx, res in enumerate(seg.residues, 1):
            res.resid = idx
        for atom in seg.atoms:
            atom.segment.segid = seg.residues[0].resname
            
    last_center = last_copy.atoms.center_of_mass()
    last_copy.atoms.positions -= last_center
    with mda.Writer(f"be{i + 1}_lastrenum.pdb", total_sel.n_atoms) as w:
        w.write(last_copy)
```

```python
data = {
    "be_1": be_1_list,
    "be_2": be_2_list,
    "gnum_dict": gnum_dict
}

with open('gnum_data.json', 'w') as f:
    json.dump(data, f, indent=4)
```
